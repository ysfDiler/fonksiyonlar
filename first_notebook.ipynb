```python
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
# Fonksiyon: İç içe döngülerle 'toplam' değişkenini artırır
def loop(N):
    toplam = 0
    for i in range(N):
        for j in range(N):
            toplam += 1
    return toplam

# Zaman karmaşıklığı ölçümü için farklı N değerlerini test eder
N_values = range(1, 501, 30)  # 1'den 500'e kadar 20'şer artırarak test ediyoruz
times = []

# Her N değeri için hesaplama süresini ölçer
for N in N_values:
    start_time = time.perf_counter()  # Daha hassas zaman ölçümü
    loop(N)
    end_time = time.perf_counter()  # Daha hassas zaman ölçümü
    times.append(end_time - start_time)

# Pandas DataFrame ile verileri oluşturur
data = pd.DataFrame({
    'N': N_values,
    'Time': times
})

# Zaman karmaşıklığı grafiği çizdirir
plt.figure(figsize=(10, 6))
plt.plot(data['N'], data['Time'], label='Time Complexity', color='g')
plt.xlabel('N (Number of elements)')
plt.ylabel('Time (seconds)')
plt.title('Time Complexity of Function (O(N^2))')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fonksiyon: 1'den N'ye kadar olan sayıların karelerinin toplamını hesaplar
def kare_toplami(N):
    toplam = 0
    for i in range(1, N + 1):
        toplam += i * i
    return toplam

# Zaman karmaşıklığı ölçümü için farklı N değerlerini test eder
N_values = range(1000, 10001, 5000)  # 1000'den 10000'e kadar 5000 artışlarla deniyoruz
times = []

# Her N değeri için hesaplama süresini ölçelim
for N in N_values:
    start_time = time.perf_counter()  # zaman ölçümü
    kare_toplami(N)
    end_time = time.perf_counter()  # zaman ölçümü
    times.append(end_time - start_time)

# Pandas DataFrame ile verileri oluştur
data = pd.DataFrame({
    'N': N_values,
    'Time': times
})

# Zaman karmaşıklığı grafiği
plt.figure(figsize=(10, 8))
plt.plot(data['N'], data['Time'], label='Time Complexity', color='b')
plt.xlabel('N (Number of elements)')
plt.ylabel('Time (seconds)')
plt.title('Time Complexity of kare_toplami Function')
plt.legend()
plt.grid(True)
plt.show()
```
